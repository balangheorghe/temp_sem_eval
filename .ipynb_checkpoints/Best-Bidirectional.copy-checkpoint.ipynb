```python
def f1(y_true, y_pred):
    y_pred = K.round(y_pred)
    tp = K.sum(K.cast(y_true*y_pred, 'float'), axis=0)
    tn = K.sum(K.cast((1-y_true)*(1-y_pred), 'float'), axis=0)
    fp = K.sum(K.cast((1-y_true)*y_pred, 'float'), axis=0)
    fn = K.sum(K.cast(y_true*(1-y_pred), 'float'), axis=0)

    p = tp / (tp + fp + K.epsilon())
    r = tp / (tp + fn + K.epsilon())

    f1 = 2*p*r / (p+r+K.epsilon())
    f1 = tf.where(tf.is_nan(f1), tf.zeros_like(f1), f1)
    return K.mean(f1)

def f1_a_p_r_s(y_true,y_pred):
    y_pred = np.round(y_pred)
    tp = np.sum(y_true*y_pred, axis=0)
    tn = np.sum((1-y_true)*(1-y_pred), axis=0)
    fp = np.sum((1-y_true)*y_pred, axis=0)
    fn = np.sum(y_true*(1-y_pred), axis=0)

    a = (tp + tn) / (len(y_true) + np.finfo(float).eps)
    p = tp / (tp + fp + np.finfo(float).eps)
    r = tp / (tp + fn + np.finfo(float).eps)
    s = tn / (tn + fp + np.finfo(float).eps)

    f1 = 2*p*r / (p+r+np.finfo(float).eps)
    return f1,np.mean(f1),a,np.mean(a),p,np.mean(p),r,np.mean(r),s,np.mean(s)
```

```python
NR_WORDS = 17000
```

```python
tokenizer = Tokenizer(num_words=NR_WORDS, filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',
                                   lower=True,split=' ')

tokenizer.fit_on_texts(full_text_data)
```

```python
X_train = tokenizer.texts_to_sequences(text_data)
X_train = pad_sequences(X_train, 159)

X_test = tokenizer.texts_to_sequences(test_data)
X_test = pad_sequences(X_test, 159)

X_valid = tokenizer.texts_to_sequences(valid_data)
X_valid = pad_sequences(X_valid, 159)
```

```python
words = {"others": 0, "angry": 1, "sad":2, "happy": 3}

def one_hot_vector(word):
    y = [0,0,0,0]
    y[words[word]] = 1
    return y

Y_train = []

for idx,row in df.iterrows():
    Y_train.append(one_hot_vector(row['label']))

Y_train = np.array(Y_train)

Y_test = []

for idx,row in df_test.iterrows():
    Y_test.append(one_hot_vector(row['label']))

Y_test = np.array(Y_test)

Y_valid = []

for idx,row in df_valid.iterrows():
    Y_valid.append(one_hot_vector(row['label']))

Y_valid = np.array(Y_valid)
```

#### model

```python
embed_dim = 256
lstm_out = 128
batch_size = 64

adam = optimizers.Adam(lr=0.01)
rmsprop = optimizers.RMSprop(lr=0.0005, rho=0.9, epsilon=None, decay=0.0)
sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)

model = Sequential()
model.add(Embedding(NR_WORDS, embed_dim,input_length = X_train.shape[1]))
model.add(Bidirectional(LSTM(lstm_out, kernel_regularizer=l2(0.000001), recurrent_regularizer=l2(0.000001))))
model.add(Dropout(0.4))
model.add(Dense(4,activation='softmax'))

model.compile(loss = 'categorical_crossentropy', optimizer=rmsprop, metrics = ['accuracy', f1])
```

```python
model.summary()
```

```python
Y_train = np.array(Y_train)

history = model.fit(X_train, Y_train,
                    validation_data=(X_valid, Y_valid),
                    epochs=20, verbose=1, batch_size=batch_size)
```

#### Evaluation

```python
Y_predict = model.predict(X_test, verbose = 2, batch_size = batch_size)

print(f1_a_p_r_s(Y_test, Y_predict))
```
